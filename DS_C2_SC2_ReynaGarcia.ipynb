```python
########### Reto - Análisis del progreso mundial de vacunación ###########

########### Por: Reyna ISabel García Toledo #############################


#P1 Extraer la información

import pandas as pd #Se obtiene liberías
vac=pd.read_csv("country_vaccinations.csv")
print(vac)

#P2 Estructura de datos y normalización de datos tipo
# datetime64

vac.info()

vac["date"]=pd.to_datetime(vac["date"])
vac.info()

#P3 Determinar la cantidad de vacunas aplicadas a cada compañía
#columna vacciones 

vac[["vaccines","daily_vaccinations"]].groupby("vaccines").sum()
 
#P4 Obtener la cantidad de vacunas aplicadas en todo el mundo

print("Total de vacunas en el mundo:",vac["daily_vaccinations"].sum())

#Total de vacunas en el mundo: 11320239871.0

#P5 Calcular el promedio de vacunas aplicadas por pais

vac[["daily_vaccinations","country"]].groupby("country").mean()

#P6 Determinar la cantidad de vacunas aplicadas en día 
#29/01/21 en todo el mundo 

vac_day=vac[vac["date"].isin(["29/01/21"])] 
print("El total de vacunas aplicadas el 29-01-21 fue",
      vac_day["daily_vaccinations"].sum())

#El total de vacunas aplicadas el 29-01-21 fue 4884052.0

#P7 Crear datagrame nuevo denominado conDiferencias
#datos origninales y derivada (diferencias)
# daily_vaaacionatios y daily_vaccionations_raw

conDiferencias=vac.copy()
conDiferencias["diferencias"]=conDiferencias["daily_vaccinations"]
-conDiferencias["daily_vaccinations_raw"]
conDiferencias

#P8 Obtener periodo de tiempo entre el resgistro con fecha más reciente
# y el registro con fecha más antigua

vac["date"].max()-vac["date"].min()
#Timedelta('482 days 00:00:00')

#P9 Crear un dataframe  nuevo denominado conCantidad con datos originales
#con columna derivada (canVac) 

conCantidad=vac.copy()
conCantidad["canVac"]=conCantidad["vaccines"].str.split("/").str.len()
conCantidad

#P10 Generar dataframe denominado antes20 con todos los registros
#que se hayan realizado antes del 20 de diciembre de 2020

antes20 = vac.loc[vac["date"]<"20/12/2020"]
antes20

#P11 Obtener dataframe denominado pfizer con todos los registros
#donde se haya utilizado la vacuna Pfizer

Pfizer= vac[vac['vaccines'].str.contains('pfizer', case=False)]
Pfizer

#P12 Alamacenar dataframe generados en excel denominado
#resultadosReto.xlsx

with pd.ExcelWriter("resultadoReto.xlsx") as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias")  
    conCantidad.to_excel(writer, sheet_name="conCantidad")
    antes20.to_excel(writer, sheet_name="antes20")
    Pfizer.to_excel(writer, sheet_name="Pfizer")  
   
#Final del Reto 

```
